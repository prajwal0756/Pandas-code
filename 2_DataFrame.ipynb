DataFrame

```python
import pandas as pd


l= [1,2,3,4,5,6]
var = pd.DataFrame(l)

print(type(var))
```

```python
d = {"a":[1,2,3,4,5], "s":[1,2,3,4,5]}

var1 = pd.DataFrame(d, columns=["a"], index=["a","b", "c","d","e"])
print(var1)
print(type(var1))
print(var1["a"][3])
```

```python
list = [[1,2,3,4],[5,6,7,8],[0,9,8,2,3]]
var2 = pd.DataFrame(list)
print(var2)
print(type(var2))
```

```python
sr = {"p":pd.Series([1,2,3,4]), "m":pd.Series([5,6,7,8])}
var3 = pd.DataFrame(sr)
print(var3)
print(type(var3))
```
